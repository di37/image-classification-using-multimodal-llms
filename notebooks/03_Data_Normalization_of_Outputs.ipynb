**Clean-up and Normalization for Ollama (Local) Models Only**

In the evaluation process, several Ollama-based local models (`minicpm-v`, `llava-llama3`, `llava`, and `llava:13b`) returned outputs that were not standardized. Issues included:

1. **Misspellings and Typos:**  
   For example, “peelecaniformes” instead of “pelecaniformes.”

2. **Verbose or Unspaced Outputs:**  
   Some predictions appeared as long descriptive sentences joined together, making the intended label unclear at first glance.

3. **Single-Letter Labels:**  
   Certain predictions consisted of only the first letter of a label, such as “p” instead of “penguin.”

To ensure a fair evaluation of these Ollama (local) models, we applied a series of normalization steps:

- **Correcting Misspellings:** Obvious misspellings of known labels were replaced with the correct forms.
- **Extracting Correct Labels from Verbose Outputs:** Long descriptive strings were trimmed down to the correct animal label.
- **Mapping Single-Letter Outputs:** If a single-letter output could be clearly matched to the correct label (based on the known ground truth), it was replaced accordingly.

**No Normalization Needed for Gemini and OpenAI Models**

In contrast, the Gemini and OpenAI models did not produce outputs requiring these adjustments. Their results were already formatted consistently, spelled correctly, and aligned with the expected label structure. Therefore, no special cleanup or normalization was necessary before evaluating their performance.

###############################################################################

Normalization Steps

Below, we apply normalization rules to ensure a fair evaluation.
These rules have been defined beforehand:
1. Correct obvious misspellings of known labels or descriptions with name of animals which we will see without spaces.
2. Normalize single-letter predictions if they unambiguously match the known animal label for that row.
3. If a correct label cannot be inferred, treat as misclassification.

###############################################################################

Normalizing `minicpm-v` results

Example: Some rows contain a long string without spaces or slightly misspelled labels like "pelecaniformes."

```python
results_ollama[results_ollama['minicpm-v_result'] == "theanimalintheimagecanbeclassifiedaspelecaniformesthesebirdsarepartofagroupthatincludespelicansandstorkswhichisdistinctenoughtowarrantitsownclassificationseparatefromotherbirdspecieslikecatsordogsthereforebasedonprovidedcategoriespelecaniformesbestfitsthisdescription"]
```

Result for this point for `minicpm-v` is `pelecaniformes` as we can see it as a sentence without spaces. Spaces has been removed during initial preprpocessing before final output is returned.

```python
results_ollama.at[124, 'minicpm-v_result'] = 'pelecaniformes'

# Also updating llava_result as we can see there is spelling mistake
results_ollama.at[124, 'llava_result'] = 'pelecaniformes'
```

```python
results_ollama[results_ollama['minicpm-v_result'] == "pelicanformes"]
```

```python
results_ollama.at[105, 'minicpm-v_result'] = 'pelecaniformes'
results_ollama.at[105, 'llava_result'] = 'pelecaniformes'
```

Normalizing `llava-llama3` results

Single-letter outputs that can be unambiguously mapped to known labels:
`p` -> `pelecaniformes` or `penguin` depending on the actual animal in that row.

```python
results_ollama[results_ollama['llava-llama3_result'] == "p"]
```

```python
results_ollama[(results_ollama['llava-llama3_result'] == "p") & (results_ollama['animal'] == "pelecaniformes")]
```

```python
results_ollama[(results_ollama['llava-llama3_result'] == "p") & (results_ollama['animal'] == "penguin")]
```

```python
mask = (results_ollama['llava-llama3_result'] == "p") & (results_ollama['animal'] == "pelecaniformes")
results_ollama.loc[mask, 'llava-llama3_result'] = 'pelecaniformes'

mask = (results_ollama['llava-llama3_result'] == "p") & (results_ollama['animal'] == "penguin")
results_ollama.loc[mask, 'llava-llama3_result'] = 'penguin'
```

```python
results_ollama[results_ollama['llava-llama3_result'] == "b"]
```
